```python
with open('ggg.csv', mode='r') as infile, open('ggg_with_addr.csv', mode='w', newline='') as outfile:
    reader = csv.DictReader(infile)
    fieldnames = reader.fieldnames + ['addr']  # Add 'addr' to the field names
    writer = csv.DictWriter(outfile, fieldnames=fieldnames)
    
    writer.writeheader()
    for row in reader:
        lat = row['lat']
        lng = row['lng']

        reverse_geocode_result = gmaps.reverse_geocode((lat, lng), language='ko')
        address = reverse_geocode_result[0]['formatted_address']
        row['addr'] = address
        writer.writerow(row)
```
